```python
# Training data
from sklearn.model_selection import train_test_split

X = data.drop(['median_house_value'], axis=1)
y = data['median_house_value']
```

<h3>Observing Corrrelations</h3>

In this housing price dataset, I've observed strong positive correlations between the total number of rooms (total_rooms), total bedrooms (total_bedrooms), population, and households. This indicates that larger houses tend to have more bedrooms, accommodate more people, and house more households. These correlations can be valuable features for predicting housing prices, but it's crucial to be aware of potential multicollinearity issues due to their strong interrelatedness. Further analysis and modeling techniques may be needed to address these relationships and build accurate predictive models.

**Since the histograms we saw previosly were skewed, we can take the logarithms of the data to visualize them more accurately.**

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize=(17,8))
```

```python
train_data.ocean_proximity.value_counts()
```

```python
pd.get_dummies(train_data.ocean_proximity)
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)
train_data
```

```python
plt.figure(figsize=(17,8))
sns.heatmap(train_data.corr(), annot=True, cmap='RdPu_r')
```

**We can see here that there is a negative correlation with 'INLAND' and 'median_house_value' of -0.49. Houses near the ocean tend to command higher prices than those located inland primarily due to the strong positive correlation between coastal proximity and property values. We can visualize this with a scatterplot as shown below**

```python
plt.figure(figsize=(17, 8))
sns.scatterplot(x='latitude', y='longitude', data=train_data, hue='median_house_value', palette='RdYlGn')
```

**We want to see if the bedroom ratio and the total household rooms would yield more interesting insights of their correlations to other attributes. Doing this shows us that there seems be some correlation with the bedroom ration and household rooms with population and median income.**

```python
train_data['bedroom_ration'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
plt.figure(figsize=(17,8))
sns.heatmap(train_data.corr(), annot=True, cmap='RdPu_r')
```

<h3>Linear Regression:</h3>

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() # Scaling data

X_train, y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)

reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
test_data = X_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)

test_data['bedroom_ration'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']
```

```python
X_test, y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
X_test_s = scaler.transform(X_test)
```

```python
reg.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor


forest = RandomForestRegressor()

forest.fit(X_train, y_train)
forest.score(X_test, y_test)
```

```python
forest.fit(X_train_s, y_train)
forest.score(X_test_s, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    'n_estimators': [5, 30, 50, 100],
    'max_features': [4, 6, 8, 20],
    'min_samples_split': [2, 4, 6, 8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)

grid_search.fit(X_train_s, y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(X_test_s, y_test)
```
